```python
import pandas as pd
```

```python
dataset = pd.read_excel("books2.xlsx")
```

```python



# Filter the dataset to include only the Harry Potter series
harry_potter_books = dataset[dataset['original_title'].str.contains('Harry Potter')]

# Print the filtered dataset
print("Filtered dataset with only Harry Potter books:")
print(harry_potter_books)

```

```python
# Data cleaning and preprocessing
# Remove duplicate rows
books_dataset = dataset.drop_duplicates()

# Drop rows with missing values
books_dataset = dataset.dropna()

# Find the most selling books within the Harry Potter series
most_selling_books = harry_potter_books.sort_values(by='books_count', ascending=False).head(5)

# Calculate the average rating of the Harry Potter books
average_rating = harry_potter_books['average_rating'].mean()

# Print the most selling books
print("Most selling books within the Harry Potter series:")
print(most_selling_books[['title', 'books_count']])

# Print the average rating
print("\nAverage rating of the Harry Potter books:", average_rating)
```
